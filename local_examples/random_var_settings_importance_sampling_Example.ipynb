```python
import sys
sys.path.append(r'C:\Users\rezio\OneDrive\Documentos\.git codes\parepy')

from parepy_toolbox import sampling_algorithm_structural_analysis
```

```python
def obj(x):
    # G(x) = 5 - x[0] - x[1] → falha quando G(x) <= 0
    return [5 - x[0] - x[1]]

# Distribuição "real" do problema
x1_real = {'type': 'normal', 'parameters': {'mean': 2.0, 'std': 1.0}}
x2_real = {'type': 'normal', 'parameters': {'mean': 2.0, 'std': 1.0}}
random_vars = [x1_real, x2_real]

# Distribuição usada no Importance Sampling (mudança de média)
x1_is = {'type': 'normal', 'parameters': {'mean': 3.0, 'std': 1.0}}
x2_is = {'type': 'normal', 'parameters': {'mean': 3.0, 'std': 1.0}}
random_vars_is = [x1_is, x2_is]

df, pf, beta = sampling_algorithm_structural_analysis(
    obj=obj,
    random_var_settings=random_vars,
    method='lhs',
    n_samples=10000,
    number_of_limit_functions=1,
    parallel=False,
    verbose=False,
    random_var_settings_importance_sampling=random_vars_is
)
```

```python
print("Probability of Failure (Pf):")
print(pf)

print("\nReliability Index (Beta):")
print(beta)

print("\nPrimeiras amostras:")
df.head()
```
