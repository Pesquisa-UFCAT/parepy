```python
# book e-Design Computer-Aided Engineering Design, Kuang-Hua Chang, https://www.sciencedirect.com/book/9780123820389/e-design, page 558 ok 
#nome chang-558
def obj(x):
    return 12.5 * x[0] ** 3 - x[1]

d = {'type': 'normal', 'parameters': {'mean': 1., 'std': 0.1}}
l = {'type': 'normal', 'parameters': {'mean': 10., 'std': 1.}}
var = [d, l]
x0 = [1, 10]
max_iter = 10
tol = 1E-8
r, pf, beta = deterministic_algorithm_structural_analysis(obj, tol, max_iter, var, x0)
print(f"pf: {pf}")
print(f"beta: {beta}")
r
```

```python
# book Segurança estrutural: Uma introdução com aplicações à segurança, Luciano Jacinto, page 155 
# none do problema Jacinto-p155
def obj(x):
    g, q, fy = x
    a_s = 4 * 0.79*1E-4
    return fy * a_s - 6.75 * (g + q)
g = {'type': 'normal', 'parameters': {'mean': 12., 'std': 0.05 * 12.}}
q = {'type': 'gumbel max', 'parameters': {'mean': 6., 'std': 0.10 * 6.}}
fy = {'type': 'normal', 'parameters': {'mean': 560E3, 'std': 0.05 * 560E3}}
var = [g, q, fy]
x0 = [12., 6., 560E3]
max_iter = 20
tol = 1E-6
r, pf, beta = deterministic_algorithm_structural_analysis(obj, tol, max_iter, var, x0)
print(f"pf: {pf}")
print(f"beta: {beta}")
r
```

```python
# book Reliability of Structures, A S Nowak e K R Collins, page 123
# nome NowakCollins-p123
def obj(x):
    r, q = x
    return r - q
r = {'type': 'lognormal', 'parameters': {'mean': 200, 'std': 20}}
q = {'type': 'gumbel max', 'parameters': {'mean': 100, 'std': 12}}
var = [r, q]
x0 = [200, 100]
max_iter = 10
tol = 1E-6
r, pf, beta = deterministic_algorithm_structural_analysis(obj, tol, max_iter, var, x0)
print(f"pf: {pf}")
print(f"beta: {beta}")
r
```
