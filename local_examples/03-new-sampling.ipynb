# _Normal sampling_

```python
# Dataset
model = {'model sampling': 'mcs'}
f = {'type': 'normal', 'loc': 7, 'scale': 3, 'seed': None}
varSet = [f]
size = 1000000

# Call function
r = sampling(size, len(varSet), model, varSet)

# Output details
print(f'random variables n={len(r)}: \n\n {r} \n\n type variable: {type(r)}')

# Histogram chart
chart_config = {
              'name': "figure_1",
              'height': 10,
              'width': 16, 
              'extension': 'svg',
              'dots_per_inch': 600,
              'x_axis_label': '$x_i$ values',
              'x_axis_size': 15,
              'y_axis_label': 'Frequency',
              'y_axis_size': 15,
              'axises_color': 'black',
              'labels_size': 15,
              'labels_color': 'black', 
              'chart_color': '#c44601',
              'bins': 20,
             }

# Call function
histogram_chart(dataset=r.flatten().tolist(), plot_setup=chart_config)
```

# _Gumbel minimum sampling - Stochastic_ 

```python
from parepy_toolbox import sampling

# Dataset: fixed value in time series
model = {'model sampling': 'mcs-time', 'time steps': 3}
f = {'type': 'lognormal', 'loc': 1, 'scale': 0.05, 'seed': None, 'stochastic variable': True}
varSet = [f]
size = 1000

# Call function
r = sampling(size, len(varSet), model, varSet)
print(f'random variables n={len(r)}: \n\n {r} \n\n type variable: {type(r)}')
groups = [r[r[:, 1] == i][:, 0] for i in np.unique(r[:, 1])]

# Hisotgram chart
chart_config = {
              'name': "figure_2_mcs_time_normal_stochastic",
              'height': 10,
              'width': 16, 
              'extension': 'svg',
              'dots_per_inch': 600,
              'x_axis_label': '$x_i$ values',
              'x_axis_size': 15,
              'y_axis_label': 'Frequency',
              'y_axis_size': 15,
              'axises_color': 'black',
              'labels_size': 15,
              'labels_color': 'black', 
              'chart_color': '#c44601',
              'bins': 20,
             }

# Call function
cores = ['#c44601', '#1f77b4', '#2ca02c']
for i, grups in enumerate(groups):
    chart_config = chart_config.copy()
    chart_config['name'] = f"figure_2_mcs_time_normal_stochastic_{i}"
    chart_config['chart_color'] = cores[i % len(cores)]
    histogram_chart(dataset=grups.flatten().tolist(), plot_setup=chart_config)
```
